```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline 
sns.set(style='whitegrid', palette='muted', font_scale=1.2)


df = pd.read_csv("../../data/Salary_Data.csv")
df.head()
```

```python
df.info()


df.describe(include='all')


df.isnull().sum()
```

Univariate Analysis

```python
numeric_cols = ['Age', 'Years of Experience', 'Salary']

for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
categorical_cols = ['Gender', 'Education Level', 'Job Title']

for col in categorical_cols:
    plt.figure(figsize=(14, 6))  # Make the figure wider
    sns.countplot(x=col, data=df, palette='pastel', order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=90, ha='right', fontsize=8)  # Rotate and align labels
    plt.tight_layout()  # Adjust layout to prevent label cut-off
    plt.show()

```

```python
top_n = 10
top_categories = df[col].value_counts().index[:top_n]
filtered_df = df[df[col].isin(top_categories)]

plt.figure(figsize=(10, 5))
sns.countplot(x=col, data=filtered_df, palette='pastel', order=top_categories)
plt.title(f'Top {top_n} {col}s')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Bivariate Analysis (Relationships)

Salary vs. Categorical

```python
for col in categorical_cols:
    plt.figure(figsize=(14, 6))  
    sns.boxplot(x=col, y='Salary', data=df, palette='Set2')
    plt.title(f'Salary vs {col}')
    plt.xticks(rotation=90, ha='right', fontsize=8)  
    plt.tight_layout() 
    plt.show()
```

```python
col = 'Job Title'

# Get top 10 most frequent job titles
top_categories = df[col].value_counts().index[:10]
filtered_df = df[df[col].isin(top_categories)]

plt.figure(figsize=(14, 6))
sns.boxplot(x=col, y='Salary', data=filtered_df, palette='Set2')
plt.title(f'Salary vs {col} (Top 10)')
plt.xticks(rotation=90, ha='right', fontsize=8)
plt.tight_layout()
plt.show()
```

Salary vs. Numeric

```python
for col in ['Age', 'Years of Experience']:
    plt.figure(figsize=(8, 4))
    sns.scatterplot(x=col, y='Salary', data=df, color='purple')
    plt.title(f'Salary vs {col}')
    plt.show()
```

Correlation Heatmap

```python
plt.figure(figsize=(8, 6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
